**What is Auto Correlation (자기상관)? **

-Series Correlation이라는 이름도 있음

한 signal하고 그의 지연된 버전와 비교할 때 사용하는 것이다. 

-만약 한 signal은 periodic signal이고 그 지연된 것은 period의 integer number라면 그 signal는 자체의 버전과 완벽한 correlation이 될 것이다.

-Autocorrelation은 signal processing의 과정에서 (특히 human hearing에 대한) 가장 종요한 부분이라고 함

-Python으로 활용하고 싶은면: 
1. Panda에 속한 **lagplot()** 사용함
2. Panda에 속한 **concat()** 하고 **shift()** function을 사용함
3. Correlogram을 만들고 싶은면 Statsmodel library중에 **plot_acf()** 사용함

**What is Root mean square? (RMS | 제곱평균제곱근)**

-Data의 mean absolute deviation을 계산할 때 사용하는 수학적 공식

-예를 들어서: Alternating Current이나, 전압, 전류의 값 등 평균속도을 표시할 때 실제 값의 root mean square으로 표현한다.
